# Loading and Preprocessing data

The `prepare_dataset` function is designed to split a given dataset into three distinct subsets: training, cross-validation, and test sets. This is a common approach in machine learning to ensure the model is evaluated and tested effectively.


````python

def prepare_dataset(x,y): 
    # Get 60% of the dataset as the training set. Put the remaining 40% in temporary variables.
    x_train, x_, y_train, y_ = train_test_split(x, y, test_size=0.40, random_state=80)

    # Split the 40% subset above into two: one half for cross validation and the other for the test set
    x_cv, x_test, y_cv, y_test = train_test_split(x_, y_, test_size=0.50, random_state=80)
    #cv=cross validation set
    return x_train, y_train, x_cv, y_cv, x_test, y_test
````

````python
def plot_dataset(x, y, title):
    plt.rcParams["figure.figsize"] = (12,8)
    plt.rcParams["lines.markersize"] = 12
    plt.scatter(x, y, marker='x', c='r'); 
    plt.title(title)
    plt.xlabel("x"); 
    plt.ylabel("y"); 
    plt.show()
````

````python
def sigmoid(x):
    g=1/(1+np.exp(-x))
    return g
````

## Gradient Descent

Gradient descent is an optimization algorithm used to minimize the cost function in machine learning and statistics.

### Cost Function

The cost function $( J(\mathbf{w}, b) $) measures the performance of a machine learning model for given values of the parameters $( \mathbf{w} $) (weights) and $( b $) (bias). For logistic regression, the cost function is given by:

$$ 
J(\mathbf{w}, b) = -\frac{1}{m} \sum_{i=1}^{m} \left[ y^{(i)} \log \left( f_{\mathbf{w}, b} \left( \mathbf{x}^{(i)} \right) \right) + (1 - y^{(i)}) \log \left( 1 - f_{\mathbf{w}, b} \left( \mathbf{x}^{(i)} \right) \right) \right]
$$

Where:
- $( m $) is the number of training examples.
- $( y^{(i)} $) is the actual label of the $( i $)-th training example.
- $( f_{\mathbf{w}, b}(\mathbf{x}^{(i)}) $) is the predicted probability for the $( i $)-th example.

### Gradient Descent Algorithm

The gradient descent algorithm iteratively adjusts the parameters $( \mathbf{w} $) and $( b $) to minimize the cost function. The update rules for the parameters are:

1. **Weight Update $( \mathbf{w}_j $)**

$$
\mathbf{w}_j = \mathbf{w}_j - \alpha \frac{\partial}{\partial \mathbf{w}_j} J(\mathbf{w}, b)
$$

Where:
- $( \alpha $) is the learning rate.
- $( \frac{\partial}{\partial \mathbf{w}_j} J(\mathbf{w}, b) $) is the partial derivative of the cost function with respect to the weight $( \mathbf{w}_j $).

The partial derivative is calculated as:

$$ 
\frac{\partial}{\partial \mathbf{w}_j} J(\mathbf{w}, b) = \frac{1}{m} \sum_{i=1}^{m} \left( f_{\mathbf{w}, b} \left( \mathbf{x}^{(i)} \right) - y^{(i)} \right) \mathbf{x}_j^{(i)}
$$

2. **Bias Update \( b \)**

$$
b = b - \alpha \frac{\partial}{\partial b} J(\mathbf{w}, b)
$$

Where:
- $( \frac{\partial}{\partial b} J(\mathbf{w}, b) $) is the partial derivative of the cost function with respect to the bias $( b $).

The partial derivative is calculated as:

$$ 
\frac{\partial}{\partial b} J(\mathbf{w}, b) = \frac{1}{m} \sum_{i=1}^{m} \left( f_{\mathbf{w}, b} \left( \mathbf{x}^{(i)} \right) - y^{(i)} \right)
$$

### Simultaneous Updates

The parameters $( \mathbf{w} $) and $( b $) are updated simultaneously in each iteration until the cost function converges to a minimum value or a predefined number of iterations is reached.

**Repeat until convergence**:

$$
\mathbf{w}_j = \mathbf{w}_j - \alpha \frac{\partial}{\partial \mathbf{w}_j} J(\mathbf{w}, b)
$$

$$
b = b - \alpha \frac{\partial}{\partial b} J(\mathbf{w}, b)
$$

This process continues iteratively to find the optimal values of $( \mathbf{w} $) and $( b $) that minimize the cost function.


````python
def compute_gradient_logistic(X, y, w, b): 
    """
    Computes the gradient for logistic regression 
 
    Args:
      X (ndarray (m,1): Data, m examples with one feature
      y (ndarray (m,)): target values
      w (ndarray (n,)): model parameters: Here since n=1 it's a scalar 
      b (scalar)      : model parameter
    Returns
      dj_dw (ndarray (n,)): The gradient of the cost w.r.t. the parameters w. n=1
      dj_db (scalar)      : The gradient of the cost w.r.t. the parameter b. 
    """
    m = X.shape[0]
    dj_dw = 0                          
    dj_db = 0

    for i in range(m):
        f_wb_i = sigmoid(np.dot(X[i],w) + b)          
        err_i  = f_wb_i  - y[i]                       
        dj_dw = dj_dw + err_i * X[i]                  
        dj_db = dj_db + err_i
    dj_dw = dj_dw/m                                   
    dj_db = dj_db/m                                   
        
    return dj_db, dj_dw  
````

````python

def compute_cost_logistic(X, y, w, b):
    """
    Computes cost using logistic loss, non-matrix version

    Args:
      X (ndarray): Shape (m,n)  matrix of examples with n features. Here n=1
      y (ndarray): Shape (m,)   target values
      w (ndarray): Shape (n,)   parameters for prediction. n=1
      b (scalar):               parameter  for prediction
      lambda_ : (scalar, float) Controls amount of regularization, 0 = no regularization
      safe : (boolean)          True-selects under/overflow safe algorithm
    Returns:
      cost (scalar): cost
    """

    m = X.shape[0]
    cost = 0.0
    for i in range(m):
        z_i    = X[i]*w + b                                            
        f_wb_i = sigmoid(z_i)                                                  
        cost  += -y[i] * np.log(f_wb_i) - (1 - y[i]) * np.log(1 - f_wb_i)     # cost function for logistic regression
    cost = cost/m

    return cost
````

````python
def gradient_descent(X, y, w_in, b_in, alpha, num_iters): 
    """
    Performs batch gradient descent
    
    Args:
      X (ndarray (m,n)   : Data, m examples with n features
      y (ndarray (m,))   : target values
      w_in (ndarray (n,)): Initial values of model parameters  
      b_in (scalar)      : Initial values of model parameter
      alpha (float)      : Learning rate
      num_iters (scalar) : number of iterations to run gradient descent
      
    Returns:
      w (ndarray (n,))   : Updated values of parameters
      b (scalar)         : Updated value of parameter 
    """
    # An array to store cost J and w's at each iteration primarily for graphing later
    J_history = []
    w = copy.deepcopy(w_in)  #avoid modifying global w within function
    b = b_in
    
    for i in range(num_iters):
        # Calculate the gradient and update the parameters
        dj_db, dj_dw = compute_gradient_logistic(X, y, w, b)   

        # Update Parameters using w, b, alpha and gradient
        w = w - alpha * dj_dw               
        b = b - alpha * dj_db               
      
        # Save cost J at each iteration
        if i<100000:      # prevent resource exhaustion 
            J_history.append( compute_cost_logistic(X, y, w, b) )

        # Print cost every at intervals 10 times or as many iterations if < 10
        if (i % 100) == 0:
         print(f"Iteration {i:4d}: Cost {J_history[-1]}   ")
        
    return w, b, J_history         #return final w,b and J history for graphing
````

````python
# dataset
# Load the CSV file
file_path = 'shuffled_file.csv' # Update the path with the correct CSV file path
df = pd.read_csv(file_path)
x = df['Tool wear [min]'].to_numpy()
x = np.array(x, dtype=np.float64)
x = x.reshape(-1, 1)
y = df['Machine failure'].to_numpy()
y = np.array(y, dtype=np.float64)
# Initialize the MinMaxScaler
scaler = MinMaxScaler()
# Apply Min-Max Scaling to the selected features
x_scaled = scaler.fit_transform(x)
# Reshape x_scaled back to 1D
x = x_scaled.flatten()

x_train, y_train, x_cv, y_cv, x_test, y_test=prepare_dataset(x,y)
````

````python
def learningcurve(x,y):
    w_tmp  = 0.1
    b_tmp  = 0.1
    alpha = 0.1
    iters = 1000
    train_mses=[]
    cv_mses=[]
    accuracys=[]
    train_sizes = np.linspace(0.1, 1.0, 10)

    for i in train_sizes:
        # Determine subset size
        subset_size = int(i * len(x))
        
        # Subset the training data
        x_subset = x[:subset_size]
        y_subset = y[:subset_size]
        x_train, y_train, x_cv, y_cv, x_test, y_test=prepare_dataset(x_subset,y_subset)
        print(f"training subset {int(10*i)}")
        w_out, b_out,train_cost_history = gradient_descent(x_train, y_train, w_tmp,  b_tmp, alpha,iters)
        train_cost=train_cost_history[-1]
        train_mses.append(train_cost)
        cv_mse = compute_cost_logistic(x_cv,y_cv,w_out,b_out)
        print(f"Cross Validation Mean Square Error : {cv_mse}")
        cv_mses.append(cv_mse)
        accuracy = Accuracy(x_test,y_test)
        print(f"Accuracy = {accuracy}")
        accuracys.append(accuracy)
    return train_mses,cv_mses,accuracys
        
````
